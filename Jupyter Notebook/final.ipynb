## Download data

In order to get data, we have to clean the provided datas and get the download url we need.

The steps are as follow:

1. get one sample url(Specimen Page URL column)
2. organize other urls with the format in 1
3. download data

```
# get url
html_data = requests.get('https://macaulaylibrary.org/asset/163678941')
soup_data = BeautifulSoup(html_data.content, 'html.parser')
soup_data
```

```
# add column according to the url: https://test.cdn.download.ams.birds.cornell.edu/api/v1/asset/163678941/
Cassin_Finch['Image_Download_URL'] = Cassin_Finch['ML Catalog Number'].apply(lambda x : str('https://test.cdn.download.ams.birds.cornell.edu/api/v1/asset/')+str(x))
House_Finch['Image_Download_URL'] = House_Finch['ML Catalog Number'].apply(lambda x : str('https://test.cdn.download.ams.birds.cornell.edu/api/v1/asset/')+str(x))
Purple_Finch['Image_Download_URL'] = Purple_Finch['ML Catalog Number'].apply(lambda x : str('https://test.cdn.download.ams.birds.cornell.edu/api/v1/asset/')+str(x))
```

```
# download
def download(bird_list, bird_name):
  i = 0
  for url in bird_list:
    ua = UserAgent()
    headers = {'User-Agent': ua.random}
    filename = "birds/"+bird_name+"/"+str(url.split("/")[6])+'.jpg'
    i += 1
    print(bird_name+"_"+str(i)+'.jpg')
    file=open(filename,'wb')
    req_img=requests.get(url,headers=headers)
    file.write(req_img.content)
    file.close()
```

```
# train sets and validation sets
Cassin_train = Cassin_Finch[0:100]
Cassin_validation = Cassin_Finch[100:]
House_train = House_Finch[0:100]
House_validation = House_Finch[100:]
Purple_train = Purple_Finch[0:100]
Purple_validation = Purple_Finch[100:]
```

```
# download images
download(Cassin_train['Image_Download_URL'], "Cassin_train")
download(Cassin_validation['Image_Download_URL'], "Cassin_validation")
download(House_train['Image_Download_URL'], "House_train")
download(House_validation['Image_Download_URL'], "House_validation")
download(Purple_train['Image_Download_URL'], "Purple_train")
download(Purple_validation['Image_Download_URL'], "Purple_validation")
```
